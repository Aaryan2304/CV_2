```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
from skimage.io import imread
from skimage.io import imsave
from skimage.io import imshow
```

```python
image = cv2.imread('IMG1.jpg')
plt.imshow(image)
```

```python
ret, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY) # threshold the image with 220 as the threshold value 
plt.imshow(mask) # show the thresholded image
kernel = np.ones((8, 8), np.uint8) 
```

```python
dilation = cv2.dilate(mask, kernel, iterations = 1) # dilate the image 
plt.imshow(dilation) 
```

```python
erosion = cv2.erode(mask, kernel, iterations = 3)
plt.imshow(erosion)
```

```python
dilation2 = cv2.dilate(erosion, kernel, iterations = 1)
plt.imshow(dilation2)
```
